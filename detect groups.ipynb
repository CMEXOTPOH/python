```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline


def get_sample(side_min=5, side_max=21, return_mask=False):
    side = np.random.randint(side_min, side_max)
    # generate matrix of ones and zeros using list comprehension
    image = np.array([[np.random.randint(0, 2) for i in range(side)] for j in range(side)])

    if return_mask:
        mask = np.zeros((side, side), dtype='bool')
        return image, mask
    return image


def dfs(x, y):  # just marks all ones connected with this as visited
    global visited, side
    visited[x][y] = True
    neighbours = get_neighbours(x, y, side)

    for neighbour in neighbours:
        x_curr, y_curr = neighbour[0], neighbour[1]
        if not visited[x_curr][y_curr] and img[x_curr, y_curr] == 1:
            dfs(x_curr, y_curr)


def get_neighbours(x, y, length):
    # verifies that the point does not exceed the boundaries
    acceptable = lambda (x, y): True if (0 <= x < length) and (0 <= y < length) else False
    # assumed neighbours
    may_be_neighbours = [(x - 1, y), (x + 1, y), (x, y + 1), (x, y - 1)]
    mask = map(acceptable, may_be_neighbours)
    real_neighbours = np.array(may_be_neighbours)[mask]

    return real_neighbours


def main():
    global visited, side, img
    groups_count = 0
    img, visited = get_sample(side_min=30, side_max=70, return_mask=True)  # you can change constants here to check on matrix of certain shape
    side = img.shape[0]

    for x in range(side):
        for y in range(side):
            if not visited[x][y] and img[x][y] == 1:
                groups_count += 1
                dfs(x, y)

    print groups_count, 'groups'
    plt.imshow(img, cmap='gray')
    plt.show()


if __name__ == '__main__':
    main()
```
